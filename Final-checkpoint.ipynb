```python
##Se crea la mascara para filtrar la linea de cada imagen mediante un filtrado del color negro de la linea 
import cv2 as cv
import numpy as np
captcha_image_files = glob.glob(os.path.join("captchas", "*.jpg"))

def filtrar_ruido(im): #El argumento de la funcion es la imagen a la que se quitará la linea (ruido)
    im_hsv = cv2.cvtColor(im , cv2.COLOR_BGR2HSV)
    lower_black = np.array([0,0,0]) #Se define el rango de color de la linea negra
    upper_black = np.array([50,50,50])
    mask = cv2.inRange(im_hsv , lower_black , upper_black ) #Nos quedamos solo con la linea

    #Se binariza la imagen del captcha y se le quita la mascara para obtener el captcha sin linea
    grayscaled = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    gaus = cv2.adaptiveThreshold(grayscaled , 255 , cv2.ADAPTIVE_THRESH_GAUSSIAN_C , cv2.THRESH_BINARY_INV, 255 , 15)
    filtrada = gaus - mask #La función devuelve la imagen sin la linea(ruido)
    return filtrada

def convolucionar(img):
    kernel = np.ones((2,2),np.float32)/1 #Definimos el kernel
    img_conv = cv2.filter2D(img, -1, kernel)   #Se realiza la convolución
    return img_conv
```
